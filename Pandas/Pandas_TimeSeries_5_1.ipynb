#### Date and Time: 
A series of time can be generated by 'date_range' command
Periods=Total Number of cycles 
Frequency='M'represents that series must be generated base on 'Month'
By Default pandas consider 'M'as the end of Month
Uses 'MS' as the start of the Month. 
D=Day
B=Buisness Day
H=Hours





```python
import pandas as pd 
import numpy as np 
```

```python
rng=pd.date_range('2019-01-04 12:40',periods=10,freq='MS')
```

```python
rng
```

Similarly, we can generate the time series using ‘start’ and ‘end’ parameters as below:

```python
rng=pd.date_range(start='2018 Jul 2 10:15', end='2018 July 12',freq='12H')
```

```python
rng
```

```python
len(rng)
```

Time zone can be specified for generating the series,

```python
rng=pd.date_range(start='2018 July 02 10:15',end='2018 Sep 02 10:15',freq='12H',tz='Asia/Kolkata')
```

```python
rng
```

```python
len(rng)
```

```python
type(rng[0]) # time stamp time
```

#### Periods : 
Periods represents the time span e.g. days, years, quarter or month etc, Period class in Panadas allows us to convert frequency easily 


#### Generating periods and frequency conversion:
In following code, period is generated using ‘Period’ command with frequency ‘M’. Note that, when we use ‘asfreq’
operation with ‘start’ operation the date is ‘01’ where as it is ‘31’ with ‘end’ option.

```python
pr=pd.Period('2012',freq='M')
```

```python
pr
```

```python
pr.asfreq('D','start')
```

```python
pr.asfreq('D','end')
```

```python
# Period arithmetic::: 
```

```python
pr=pd.Period('2012',freq='A')     # A for Anual
```

```python
pr
```
