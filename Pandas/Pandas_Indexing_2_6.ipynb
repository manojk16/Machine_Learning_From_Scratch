#### Index: 
    Index are used to organize the data and provide us fast access to data;
    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cast=pd.read_csv("cast.csv",index_col=None)
cast.head()
```

```python
%%time 
#data access without indexing:
cast[cast['title']=='Macbeth']
```

#### set_index:
can be used to set the index for data, In below code 'title' is set at index, So index-numbers are replaced by 'title'

```python
cast[cast['title']=='Macbeth']
c=cast.set_index('title')
c.head(4)
```

```python
%%time
c=cast.set_index(['title'])
c.head(4)
```

To use the above indexing .loc should be used for fast operation

```python
%%time
#data access with indexing

c.loc['Macbeth']
```

```python
cs=cast.set_index(['title']).sort_index()
cs.tail(4)
```

```python
%%time
cs.loc['Macbeth']
```

#Multiplex Indexing : Multiple Index in to data 

```python
cm=cast.set_index(['title','n']).sort_index()
cm.tail(5)
```

```python
cm.loc['Macbeth']
```

```python
# Show Macbeth with ranking 4-18
cm.loc['Macbeth'].loc[4:18]
```

If there is only one match data, then Series will return (instead of DataFrame

```python
cm.loc['Macbeth'].loc[4]
```
